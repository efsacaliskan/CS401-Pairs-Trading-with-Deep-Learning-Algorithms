```python
df = pd.read_csv('./sp500_2012_2022_1.csv', index_col=['symbol'])
timeline = pd.read_csv('./timeline_modified.csv')
```

```python
print(df.isnull().sum()) # NaN values
print(np.isinf(df).sum()) # Inf values
```

```python
 # replace the Inf values with NaN 
df = df.replace([np.inf, -np.inf], np.nan)
# remove the Nan values
df = df.dropna() 
df
```

```python
# remove unrealistic dates
timeline['Added'] = pd.to_datetime(timeline['Added'], errors='coerce')
timeline['Removed'] = pd.to_datetime(timeline['Removed'], errors='coerce')

# date range 
start_date = pd.to_datetime('2012-01-01')
end_date = pd.to_datetime('2022-01-10')  


# trial time range (minimaze the data)
timeline= timeline[
    (timeline['Added'] <= start_date) &  (timeline['Removed'] >= end_date)]

# Check cleaned data
timeline
```

```python
# filter the data according to the 'Ticker' names in timeline
filter_data = df.T
data_col = np.array(filter_data.columns)
ticker_name = timeline['Ticker'].values
filtered_ticker = np.intersect1d(data_col,ticker_name)
```

```python
# show filtered data
data = df.loc[filtered_ticker].T
data
```

```python
data = data.loc[:, data.nunique() > 1]
data
```

```python
for column in data.columns:
    data[column] = data[column].replace('[\.,]', '', regex=True).astype(float)

plt.figure(figsize=(15, 15))
sns.heatmap(data.corr(), annot=False)
plt.show()
```

```python
def find_cointegrated_pairs(data):
    n = data.shape[1]
    score_matrix = np.zeros((n, n))
    pvalue_matrix = np.ones((n, n))
    keys = data.keys()
    pairs = {}
    for i in range(n):
        for j in range(i+1, n):
            S1 = data[keys[i]]
            S2 = data[keys[j]]
            result = coint(S1, S2)
            score = result[0]
            pvalue = result[1]
            score_matrix[i, j] = score
            pvalue_matrix[i, j] = pvalue
            if pvalue < 0.005:
                pairs[(keys[i], keys[j])] = result
    return score_matrix, pvalue_matrix, pairs

scores, pvalues, pairs = find_cointegrated_pairs(data)
```

```python
sns.heatmap(pvalues,  cmap='RdYlGn_r' 
                , mask = (pvalues >= 0.98)
                )
plt.show()
```

```python
pairs_data = {key:value[1]  for (key, value) in pairs.items()}
pairs_data 
```

```python
pair1 = 'PCAR'
pair2 = 'WTW'
price1 = data[pair1]
price2 = data[pair2]
```

```python
plt.figure(figsize=(15,7))
plt.plot(price1, label = pair1)
plt.plot(price2, label = pair2)
plt.title('Historical Adjusted Close Prices')
plt.legend()
```

```python
plt.figure(figsize=(15,7))
plt.scatter(price1, price2)
plt.title(f'Scatter plot of {pair1} vs {pair2}')
plt.xlabel(f'{pair1} values')
plt.ylabel(f'{pair2} values')
plt.show()
```

```python
results = sm.OLS(price1, price2).fit()
predict = results.predict(price2)
            
error = price1 - predict
ADFtest = ts.adfuller(error)
```

```python
plt.figure(figsize = (10,7))
plt.title(f'ADFtest error of {pair1} and {pair2}')
plt.plot(error)
```

```python
plt.figure(figsize = (10,7))
ratio = price1 / price2
plt.plot(ratio, label = f'Price ratio of {pair1} / {pair2}')
plt.legend()
plt.title(f'Price ratio of {pair1} and {pair2}')
```

```python
# z-score using the ADFtest error
plt.figure(figsize = (10,7))
df_zscore1 = error/error.std()
plt.plot(df_zscore1, label = "Z Scores")
plt.axhline(df_zscore1.mean(), color = 'black')
plt.axhline(1.0, color='red') 
plt.axhline(1.25, color='red') 
plt.axhline(-1.25, color='green') 
plt.legend(loc = 'best')
plt.title(f'Z score of Error of {pair1} to {pair2}')
plt.show()
```

```python
# z-score using the price ratio
plt.figure(figsize = (10,7))
df_zscore = (ratio - ratio.mean())/ratio.std()
plt.plot(df_zscore, label = "Z Scores")
plt.axhline(df_zscore.mean(), color = 'black')
plt.axhline(1.0, color='red') 
plt.axhline(1.25, color='red') 
plt.axhline(-1.0, color='green') 
plt.axhline(-1.25, color='green')
plt.legend(loc = 'best')
plt.title(f'Z score of Ratio of {pair1} to {pair2}')
plt.show()
```
